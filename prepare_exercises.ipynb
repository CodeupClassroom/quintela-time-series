# Store sales item data

## Question 1

Convert date column to datetime format.

```python
df.sale_date = pd.to_datetime(df.sale_date, infer_datetime_format=True)
df.head()
```

## Question 2

Plot the distribution of sale_amount and item_price.

```python
sns.histplot(df.sale_amount)
plt.title('Distribution of sale amount')
```

```python
sns.histplot(df.item_price)
plt.title('Distribution of item price')
```

## Question 3

Set the index to be the datetime variable.

```python
df.set_index('sale_date', inplace=True)
df.head()
```

## Question 4

Add a 'month' and 'day of week' column to your dataframe.

```python
df['month'] = df.index.strftime('%B')
df.head()
```

```python
df['day_of_week'] = df.index.strftime('%A')
df.head()
```

## Question 5

Add a column to your dataframe, sales_total, which is a derived from sale_amount (total items) and item_price.

```python
df['sales_total'] = df.sale_amount * df.item_price
df.head()
```

# German power data

## Question 1

Convert date column to datetime format.

```python
df_power = ac.get_germany_data()
df_power.head()
```

```python
df_power.Date = pd.to_datetime(df_power.Date, infer_datetime_format=True)
df_power.head()
```

## Question 2

Plot the distribution of each of your variables.

```python
for col in list(df_power.columns.drop('Date')):
    plt.figure()
    sns.histplot(df_power[col])
    plt.title('Distribution of {}'.format(col))
```

## Question 3

Set the index to be the datetime variable.

```python
df_power.set_index('Date', inplace=True)
df_power.head()
```

## Question 4

Add a month and a year column to your dataframe.

```python
df_power['month'] = df_power.index.strftime('%B')
df_power['year'] = df_power.index.strftime('%Y')
df_power.head()
```
